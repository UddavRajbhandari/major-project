```python
# =============================
# LIME & SHAP Explainer for Kaggle (Nepali Font Fixed)
# =============================

import os
import numpy as np
import pandas as pd
import torch
import re
import emoji
import regex
import warnings
warnings.filterwarnings("ignore")
from typing import Optional
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties, fontManager

from transformers import AutoTokenizer, AutoModelForSequenceClassification
from huggingface_hub import hf_hub_download
import joblib

# ============================ FONT CONFIG ==================================
FONT_PATH = '/kaggle/input/fontss/Kalimati.ttf'  # Kaggle uploaded font

def load_nepali_font(font_path):
    if not os.path.exists(font_path):
        print(f"⚠️ Nepali font not found: {font_path}")
        return None
    fontManager.addfont(font_path)
    fp = FontProperties(fname=font_path)
    print(f"✓ Loaded Nepali font: {fp.get_name()}")
    return fp

NEPALI_FONT = load_nepali_font(FONT_PATH)
plt.rcParams["axes.unicode_minus"] = False

OUT_DIR = '/kaggle/working/explanations'
os.makedirs(OUT_DIR, exist_ok=True)

# ============================ LIME / SHAP ==================================
try:
    from lime.lime_text import LimeTextExplainer
    LIME_AVAILABLE = True
except ImportError:
    LIME_AVAILABLE = False
    print("⚠️ LIME not installed")

try:
    import shap
    SHAP_AVAILABLE = True
except ImportError:
    SHAP_AVAILABLE = False
    print("⚠️ SHAP not installed")

# ============================ TRANSLITERATION ===============================
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate

DIRGHIKARAN_MAP = {
    "उ": "ऊ", "इ": "ई", "ऋ": "रि", "ए": "ऐ", "अ": "आ",
    "\u200d": "", "\u200c": "", "।": ".", "॥": ".",
    "ि": "ी", "ु": "ू"
}

def is_devanagari(text):
    return isinstance(text, str) and bool(regex.search(r"\p{Devanagari}", text))

def roman_to_devanagari(text):
    try: return transliterate(text, sanscript.ITRANS, sanscript.DEVANAGARI)
    except: return text

def normalize_dirghikaran(text):
    for k, v in DIRGHIKARAN_MAP.items(): text = text.replace(k, v)
    return text

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+|#\w+", "", text)
    text = emoji.replace_emoji(text, replace="")
    text = re.sub(r"\s+", " ", text).strip()
    return text

def preprocess_for_transformer(text):
    if not is_devanagari(text):
        text = roman_to_devanagari(text)
    return normalize_dirghikaran(clean_text(text))

# ============================ CONFIG ========================================
TEST_FILE = '/kaggle/input/nepalihatee/test.json'
LOCAL_MODEL_PATH = '/kaggle/input/xlm-roberta-nepali-hate'
HF_MODEL_ID = "UDHOV/xlm-roberta-large-nepali-hate-classification"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
OUT_DIR = '/kaggle/working/explanations'
os.makedirs(OUT_DIR, exist_ok=True)
# ============================ MODEL LOADING ================================
def load_model():
    if os.path.exists(LOCAL_MODEL_PATH):
        tokenizer = AutoTokenizer.from_pretrained(LOCAL_MODEL_PATH)
        model = AutoModelForSequenceClassification.from_pretrained(LOCAL_MODEL_PATH)
        le = joblib.load(os.path.join(LOCAL_MODEL_PATH, "label_encoder.pkl"))
        print("✅ Model loaded locally")
    else:
        tokenizer = AutoTokenizer.from_pretrained(HF_MODEL_ID)
        model = AutoModelForSequenceClassification.from_pretrained(HF_MODEL_ID)
        le_path = hf_hub_download(HF_MODEL_ID, "label_encoder.pkl")
        le = joblib.load(le_path)
        print("✅ Model loaded from HuggingFace")

    model.to(DEVICE).eval()
    return model, tokenizer, le

# ============================ FONT HANDLING ==================================
def apply_nepali_font(
    ax: plt.Axes,
    nepali_font: Optional[FontProperties] = None,
    texts: Optional[list] = None,
    is_tick_labels: bool = True
) -> None:
    """
    Apply Nepali font to any Devanagari text.
    """
    if nepali_font is None:
        return

    if is_tick_labels or texts is None:
        for txt in ax.get_yticklabels():
            label_text = txt.get_text()
            if regex.search(r'\p{Devanagari}', label_text):
                txt.set_fontproperties(nepali_font)
                txt.set_fontsize(11)
    else:
        for txt in texts:
            if regex.search(r'\p{Devanagari}', txt.get_text()):
                txt.set_fontproperties(nepali_font)
                
def normalize_for_explanation(text: str) -> str:
    """
    Light normalization for explanation ONLY.
    - Preserves Romanized words
    - Preserves Devanagari surface form
    - Removes zero-width chars
    - Does NOT transliterate
    - Does NOT lowercase aggressively
    """
    if not isinstance(text, str):
        return ""

    # Remove zero-width joiners / non-joiners
    text = text.replace("\u200d", "").replace("\u200c", "")

    # Normalize punctuation spacing
    text = re.sub(r"\s+", " ", text).strip()

    return text

# ============================ TOKEN RECONSTRUCTION ==========================
def reconstruct_words_from_text(text: str):
    """
    Reconstruct words from text, supporting both Devanagari and Romanized Nepali.
    """
    norm_text = normalize_for_explanation(text)
    words = []

    # Devanagari words
    for match in regex.finditer(r"\p{Devanagari}+", norm_text):
        words.append(match.group())

    # Romanized / ASCII words
    for match in re.finditer(r"[a-zA-Z0-9]+", norm_text):
        words.append(match.group())

    return words, norm_text



def aggregate_token_attributions(
    tokens,
    token_values,
    original_text
):
    """
    Aggregate token-level attributions into word-level attributions.
    """
    words = reconstruct_words_from_text(original_text)
    word_scores = []

    for word, start, end in words:
        score = 0.0
        for tok, val in zip(tokens, token_values):
            if tok.strip() and tok in word:
                score += val
        word_scores.append((word, score))

    return word_scores

def build_token_alignment(original_text: str):
    """
    Align original tokens with normalized Devanagari/Romanized tokens.
    Returns:
      - model_tokens: normalized tokens for attribution
      - display_tokens: original surface form for plotting
    """
    original_tokens = original_text.split()
    model_tokens, display_tokens = [], []

    for tok in original_tokens:
        norm_tok = normalize_for_explanation(tok)
        if regex.search(r"\p{Devanagari}", norm_tok) or re.search(r"\w+", norm_tok):
            # keep Devanagari and alphanumeric Romanized words
            model_tokens.append(norm_tok)
            display_tokens.append(tok)

    return model_tokens, display_tokens

# ============================ MODEL WRAPPER ================================
class XLMRobertaExplainer:
    def __init__(self, model, tokenizer, label_encoder):
        self.model = model
        self.tokenizer = tokenizer
        self.class_names = label_encoder.classes_.tolist()

    def predict_proba(self, texts):
        if isinstance(texts, str):
            texts = [texts]
        texts = [preprocess_for_transformer(t) for t in texts]
        enc = self.tokenizer(
            texts, padding=True, truncation=True,
            max_length=256, return_tensors="pt"
        ).to(DEVICE)
        with torch.no_grad():
            probs = torch.softmax(self.model(**enc).logits, dim=-1)
        return probs.cpu().numpy()
    
    def predict_with_analysis(self, text):
        probs = self.predict_proba(text)[0]
        pred_idx = int(np.argmax(probs))
        return {
            "predicted_label": self.class_names[pred_idx],
            "confidence": float(probs[pred_idx]),
            "probabilities": {label: float(prob) for label, prob in zip(self.class_names, probs)}
        }

# ============================ LIME ==========================================
class LIMEExplainer:
    def __init__(self, model_explainer, nepali_font=None):
        self.model_explainer = model_explainer
        self.nepali_font = nepali_font
        self.explainer = LimeTextExplainer(
            class_names=model_explainer.class_names,
            random_state=42
        )
    
    def explain_and_visualize(self, text):
        exp = self.explainer.explain_instance(
            text,
            self.model_explainer.predict_proba,
            num_samples=200
        )
    
        token_weights = dict(exp.as_list())
    
        model_tokens, display_tokens = build_token_alignment(text)
    
        word_scores = []
        for model_tok, display_tok in zip(model_tokens, display_tokens):
            score = sum(
                val for tok, val in token_weights.items()
                if tok in model_tok
            )
            word_scores.append((display_tok, score))  # 👈 DISPLAY ORIGINAL
    
        if not word_scores:
            print("⚠️ No explainable words found")
            return
    
        features, weights = zip(*word_scores)
        y_pos = range(len(features))
    
        fig, ax = plt.subplots(figsize=(8, 6))
        ax.barh(y_pos, weights)
        ax.set_yticks(y_pos)
        ax.set_yticklabels(features, fontsize=11)
        ax.invert_yaxis()
        ax.set_xlabel("Contribution")
    
        # Apply Nepali font ONLY if Devanagari
        apply_nepali_font(ax, self.nepali_font)
    
        plt.tight_layout()
        out_path = os.path.join(OUT_DIR, f"lime_explanation_{abs(hash(text)) % 10**8}.png")
        plt.savefig(out_path, dpi=300, bbox_inches="tight")
        plt.show()
        print(f"✅ LIME image saved to: {out_path}")


# ============================ SHAP ==========================================
class SHAPExplainer:
    def __init__(self, model_explainer, nepali_font=None):
        """
        SHAP explainer for Devanagari + Romanized Nepali text.
        """
        self.model_explainer = model_explainer
        self.nepali_font = nepali_font

        self.explainer = shap.Explainer(
            model_explainer.predict_proba,
            shap.maskers.Text(model_explainer.tokenizer)
        )

    def explain_and_visualize(self, text: str):
        # ---- Run SHAP ----
        sv = self.explainer([text])[0]

        tokens = list(sv.data)
        values = sv.values
        values_array = np.array(values)

        # ---- Select predicted class safely ----
        pred_probs = self.model_explainer.predict_proba([text])[0]
        class_idx = int(np.argmax(pred_probs))

        if values_array.ndim == 1:
            token_values = values_array
        elif values_array.ndim == 2:
            # shape: (tokens, classes)
            token_values = values_array[:, class_idx]
        elif values_array.ndim == 3:
            # shape: (samples, tokens, classes)
            token_values = values_array[0, :, class_idx]
        else:
            raise ValueError(f"Unexpected SHAP values shape: {values_array.shape}")

        # ---- Safety: align lengths ----
        max_len = min(len(tokens), len(token_values))
        tokens = tokens[:max_len]
        token_values = token_values[:max_len]

        # ---- Build alignment (MODEL vs DISPLAY tokens) ----
        model_tokens, display_tokens = build_token_alignment(text)
        if not display_tokens:
            print("⚠️ No explainable words found")
            return

        # ---- Aggregate token attributions to word level ----
        word_scores = []
        token_ptr = 0
        for model_tok, disp_tok in zip(model_tokens, display_tokens):
            score = 0.0
            consumed = ""
            while token_ptr < max_len and len(consumed) < len(model_tok):
                score += float(token_values[token_ptr])
                consumed += tokens[token_ptr].replace("▁", "")
                token_ptr += 1
            word_scores.append((disp_tok, score))

        if not word_scores:
            print("⚠️ Empty SHAP attribution")
            return

        # ---- Normalize for coloring ----
        max_val = max(abs(v) for _, v in word_scores) + 1e-6

        # ---- Plot ----
        fig, ax = plt.subplots(figsize=(max(8, 0.45 * len(word_scores)), 2.4))
        ax.axis("off")

        x, y = 0.01, 0.5
        text_objs = []

        for word, val in word_scores:
            intensity = min(abs(val) / max_val, 1.0)
            color = (1.0, 1.0 - intensity, 1.0 - intensity)

            txt = ax.text(
                x, y, f" {word} ",
                fontsize=13,
                bbox=dict(
                    facecolor=color,
                    edgecolor="none",
                    alpha=0.85,
                    boxstyle="round,pad=0.3"
                )
            )
            text_objs.append(txt)
            x += 0.04 + 0.028 * len(word)
            if x > 0.95:
                x = 0.01
                y -= 0.45

        # ---- Apply Nepali font ONLY to Devanagari ----
        apply_nepali_font(ax, self.nepali_font, texts=text_objs, is_tick_labels=False)

        # ---- Save plot ----
        class_name = self.model_explainer.class_names[class_idx]
        out_path = os.path.join(
            OUT_DIR,
            f"shap_explanation_{class_name}_{abs(hash(text)) % 10**8}.png"
        )
        plt.tight_layout()
        plt.savefig(out_path, dpi=300, bbox_inches="tight")
        plt.show()
        print(f"✅ SHAP image saved to: {out_path}")




# ============================ EXECUTION =====================================
if __name__=="__main__":
    model, tokenizer, le = load_model()
    explainer = XLMRobertaExplainer(model, tokenizer, le)

    df = pd.read_json(TEST_FILE)
    pd.set_option('display.max_colwidth', None)

    SAMPLES_PER_CLASS = 3
    sampled_df = df.groupby("Label_Multiclass", group_keys=False).apply(
        lambda x: x.sample(n=min(len(x), SAMPLES_PER_CLASS), random_state=42)
    )

    for i, (_, row) in enumerate(sampled_df.iterrows(), start=1):
        text = str(row["Comment"])
        true_label = row.get("Label_Multiclass", "N/A")
        analysis = explainer.predict_with_analysis(text)

        print(f"\nSample {i}/{len(sampled_df)}")
        print(f"Text: {text}")
        print(f"True Label: {true_label}")
        print(f"Prediction: {analysis['predicted_label']}")
        print(f"Confidence: {analysis['confidence']:.4f}")

        if LIME_AVAILABLE:
            print("\n--- LIME Explanation ---")
            LIMEExplainer(explainer, NEPALI_FONT).explain_and_visualize(text)

        if SHAP_AVAILABLE:
            print("\n--- SHAP Explanation ---")
            SHAPExplainer(explainer, NEPALI_FONT).explain_and_visualize(text)
```
